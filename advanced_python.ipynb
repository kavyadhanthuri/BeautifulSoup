```python
import random
p1=input("enter player1: ")
p2=input("enter player2: ")
s1=10
s2=10
d1=random.randint(1,10)
d2=random.randint(1,10)
print("_________PLAYER1 TURN________")
while(True):
   g1=int(input("guess:"))
   s1=s1-1
   if g1==d1:
      break
print("_________PLAYER2 TURN________")
while(True):
   g2=int(input("guess:"))
   s2=s2-1
   if g2==d2:
      break
print("_________SUMMARY________")
print("{}:{}".format(p1,s1))
print("{}:{}".format(p2,s2))
print("_________")
if s1>s2:
  print("winner:",p1)
elif s2>s1:

  print("winner:",p2)
else:
  print("draw")


```

```python
# trail and error method
#a poultry farm has group of hens ans cows..need to find num of cows & hens separatly, if tottal legs & total heads together given:
#testcase  1: legs:16 heads:6
#cows:2
#hens:4
#testcase 2: legs:7
#head :3 output : no solution
legs=int(input("no of legs:"))
heads=int(input("no of heads:"))
for cows in range(0,heads+1):
  hens=heads-cows
  cal_legs=cows*4 + hens*2
  if cal_legs==legs:
    print("cows",cows)
    print("hens",hens)
    break
```

```python
# trail and error method
#a poultry farm has group of hens ans cows..need to find num of cows & hens separatly, if tottal legs & total heads together given:
#testcase  1: legs:16 heads:6
#cows:2
#hens:4
#testcase 2: legs:7
#head :3 output : no solution
legs=int(input("no of legs:"))
heads=int(input("no of heads:"))
flag=False
for cows in range(0,heads+1):
  hens=heads-cows
  cal_legs=cows*4 + hens*2
  if cal_legs==legs:
    flag=True
    break
if flag:
  print("cows",cows)
  print("hens",hens)
else:
  print("no solution")
```
